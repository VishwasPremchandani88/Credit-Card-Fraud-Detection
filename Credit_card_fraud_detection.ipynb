```python
import numpy as np
import pandas as pd
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card_data= pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset infromation
credit_card_data.info()
```

```python
#checking the no of missing values
credit_card_data.isna().sum()
```

```python
credit_card_data.dropna(inplace=True)
```

```python
credit_card_data.isna().sum()
```

```python
#distribution of legit transactions & fradulaent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0--> Normal Transaction
1--> Fradulent Transaction

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions


No. of Fradulent Transaction -->188



```python
legit_sample = legit.sample(n=188)
```

Concatenating two DataFrames


```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets


```python
X= new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and testing data


```python
X_train,X_test,Y_train,Y_test =  train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training


Logistic Regression


```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
```

Model Evaluation


Accuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on the training data =",training_data_accuracy)
```

```python
#accuracy on test on data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on the training data =",test_data_accuracy)
```
